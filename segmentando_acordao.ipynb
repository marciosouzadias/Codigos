# Segmentando com a NLTK:

```python
class SegmenterNLTK():

  def __init__(self):

    self.predicted_sentences = []
    self.number_predicted_sentences = 0


  def segment_with_nltk(self, text):
    
    # NOTE: segmenting with NLTK the original texts in sentences:
    self.predicted_sentences.extend(sent_tokenize(text, language="portuguese"))

    for idx, sentence in enumerate(self.predicted_sentences):
      print(f"sentença número {idx}: \n {sentence} \n")
      
    
  def number_predicted(self):
    
    print(f"number of predicted sentences by NLTK: {len(self.predicted_sentences)}")

    
    
```

```python
nltk_segmenter = SegmenterNLTK()
```

```python
nltk_segmenter.segment_with_nltk(texto_acordao)
```

```python
nltk_segmenter.number_predicted()
```

# Segmentando com o LexNLP:

```python
class SegmenterLexNLP():

    def __init__(self):

      self.predicted_sentences = []

    
    def segment_with_lexnlp(self, text):
      
      # NOTE: segmenting with LexNLP the original texts in sentences:
      self.predicted_sentences.extend(lexnlp.get_sentence_list(text))


      print(f"number of predicted sentences by LexNLP: {len(self.predicted_sentences)}")

          
      for idx, sentence in enumerate(self.predicted_sentences):
        print(f"sentença número {idx}: \n {sentence} \n")


    def number_predicted(self):
    
      print(f"number of predicted sentences by LexNLP: {len(self.predicted_sentences)}")
```

```python
lexnlp_segmenter = SegmenterLexNLP()
```

```python
lexnlp_segmenter.segment_with_lexnlp(texto_acordao)
```

```python
lexnlp_segmenter.number_predicted()
```

# Segmentando com a SpaCy:

```python
class SegmenterSpacy():

    def __init__(self):
      self.predicted_sentences = []
    
    def segment_with_spacy(self, text):

      spacy_model = spacy.load("pt_core_news_lg") # Load the large Portuguese model
      
      # NOTE: generating a single list with all the segmented sentences by SpaCy: 
      self.predicted_sentences.extend(str(sent) for sent in spacy_model(text).sents)

      for idx, sentence in enumerate(self.predicted_sentences):
        print(f"sentença número {idx}: \n {sentence} \n")

    
    def number_predicted(self):
    
      print(f"number of predicted sentences by SpaCy: {len(self.predicted_sentences)}")
```

```python
spacy_segmenter = SegmenterSpacy()
```

```python
spacy_segmenter.segment_with_spacy(texto_acordao)
```

```python
spacy_segmenter.number_predicted()
```

# Segmentando com o Ulysses Segmenter:

Note que mais configurações são possíveis, como utilizar outro tipo de pooling, ou ainda, não remover os noisy segments, ou usar outro modelo (LSTM):

```python
class SegmenterUlysses():

    def __init__(self):

      self.predicted_sentences = []

      self.segmenter_bert = segmentador.BERTSegmenter( device="cpu",  # or 'cuda' for GPU
                                                  inference_pooling_operation="assymetric-max",
                                                )


    
    def segment_with_ulysses(self, text):
      
      # NOTE: segmenting with Ulysses the original texts in sentences:

      seg_result = self.segmenter_bert(text,
                                  return_logits = True,
                                  remove_noise_subsegments = True
                                )
  
      self.predicted_sentences.extend(seg_result.segments)

          
      for idx, sentence in enumerate(seg_result.segments):
        print(f"segmento número {idx}: \n {sentence} \n")


    def number_predicted(self):
    
      print(f"number of predicted segments by Ulysses: {len(self.predicted_sentences)}")
```

```python
ulysses_segmenter = SegmenterUlysses()
```

```python
ulysses_segmenter.segment_with_ulysses(texto_acordao)
```

```python
ulysses_segmenter.number_predicted()
```
